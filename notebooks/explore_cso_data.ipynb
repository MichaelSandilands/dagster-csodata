# Google Cloud Storage Check

If `dlt` worked properly, we should be able to extract the file from our bucket.

```python
import os
import pandas as pd
import gcsfs
from dotenv import load_dotenv, find_dotenv

# 1. Load the .env file
load_dotenv(find_dotenv())

credentials_path = os.getenv("GOOGLE_APPLICATION_CREDENTIALS")
bucket_url = os.getenv("DESTINATION__FILESYSTEM__BUCKET_URL")

print("Credentials path from .env:", credentials_path)
print("Base Bucket URL:", bucket_url)

# 2. SAFETY CHECK: Ensure the JSON file actually exists
if not credentials_path or not os.path.exists(credentials_path):
    raise FileNotFoundError(f"Could not find the credentials file at: {credentials_path}. Please check your .env file!")

# 3. Initialize GCS File System
fs = gcsfs.GCSFileSystem(token=credentials_path)

# 4. Find the active file
dataset_name = "cso_cube_data"
table_id = "cpm02" # <-- lowercase!

target_folder = f"{bucket_url}/{dataset_name}/{table_id}"
print(f"Searching in: {target_folder}")

# The rest of your cell...
files = fs.ls(target_folder)
parquet_files = [f for f in files if f.endswith('.parquet')]
active_file_path = parquet_files[0]

# 5. Read into pandas
df = pd.read_parquet(f"gcs://{active_file_path}", storage_options={"token": credentials_path})
df.head()
```
